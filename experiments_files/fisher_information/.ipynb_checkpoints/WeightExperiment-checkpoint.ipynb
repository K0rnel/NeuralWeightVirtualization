```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
def vectorize_list(list_to_vectorize):
    vector_list = []
    for item in list_to_vectorize:
        vector_list.append(item.flatten())
    return np.concatenate(vector_list)
```

```python
def get_weights_and_fisher(name): 
    weight_filename = name + '_weight.npy'
    weight_filepath = os.path.join("./", weight_filename)
    weight = np.load(weight_filepath, allow_pickle=True, encoding="latin1")
    
    fisher_filename = name + '_fisher.npy'
    fisher_filepath = os.path.join("./", fisher_filename)
    fisher = np.load(fisher_filepath, allow_pickle=True, encoding="latin1")
    
    w = vectorize_list(weight)
    f = vectorize_list(fisher)
    
    return (w, f)
```

```python
w, f = get_weights_and_fisher("cifar10")
```

```python
len(f)
```

```python
plt.scatter(w, f)
plt.show()
```

```python
def draw_hist(name): 
    weight_filename = name + '_weight.npy'
    weight_filepath = os.path.join("./", weight_filename)
    weight = np.load(weight_filepath, allow_pickle=True, encoding="latin1")
    
    fisher_filename = name + '_network_fisher.npy'
    fisher_filepath = os.path.join("/Users/samuilstoychev/Desktop/AFTER_5/NeuralWeightVirtualization/" + name, 
                               fisher_filename)
    fisher = np.load(fisher_filepath, allow_pickle=True, encoding="latin1")
    
    plt.scatter(vectorize_list(weight), vectorize_list(fisher), bins=2000)
    plt.show()
```
