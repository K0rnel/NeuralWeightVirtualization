```python
def get_accuracy(out): 
    accs = []
    for x in out: 
        if "Inference" in x: 
            accuracy = float(x.split(":")[1])
            accs.append(accuracy)
    return accs
```

```python
### Auxiliary function: Extract information from execution logs
```

```python
def get_exec_time(out): 
    weight_load_times = []
    dnn_exec_times = [] 
    total_weight_load_time = None
    total_dnn_exec_time = None
    for x in out: 
        if "weights load time" in x: 
            weight_load_time = float(x.split(":")[1][:-3])
            if "total" not in x: 
                weight_load_times.append(weight_load_time)
            else: 
                total_weight_load_time = weight_load_time
        elif "DNN execution time" in x: 
            dnn_exec_time = float(x.split(":")[1][:-3])
            if "total" not in x: 
                dnn_exec_times.append(dnn_exec_time)
            else: 
                total_dnn_exec_time = dnn_exec_time
    return (weight_load_times, dnn_exec_times, total_weight_load_time, total_dnn_exec_time)
```

# Execution Time (3 networks)

```python
BASELINE_FILE = "Group Project - ML Systems/experiment_files/three_nets/baseline_execution_three_nets.txt"
IN_MEMORY_FILE = "Group Project - ML Systems/experiment_files/three_nets/in_memory_execution_three_nets.txt"

baseline = open(BASELINE_FILE, "r").read().splitlines()
in_memory = open(IN_MEMORY_FILE, "r").read().splitlines()

bas = get_exec_time(baseline)
inm = get_exec_time(in_memory)

print("Baseline DNN execution time:", bas[3])
print("In-memory DNN execution time:", inm[3])

### DNN Execution Time

plt.plot(bas[1], label="baseline ")
plt.plot(inm[1], label="in-memory")
plt.legend()
plt.show()

### Weights Load Time

print("Baseline weights load time:", bas[2])
print("In-memory weights load time:", inm[2])

plt.plot(bas[0], label="baseline")
plt.plot(inm[0], label="in-memory")
plt.legend()
plt.show()
```
